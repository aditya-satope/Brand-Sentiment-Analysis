```python
tweet_df = pd.read_excel('Development Data/dev_data_tweet.xlsx')
```

```python
tweet_df_train_dev, tweet_df_test = train_test_split(tweet_df,stratify=tweet_df['Mobile_Tech_Tag'],random_state=21,test_size = 0.1)
tweet_df_train, tweet_df_dev  = train_test_split(tweet_df_train_dev,stratify=tweet_df_train_dev['Mobile_Tech_Tag'],random_state=21,test_size = 0.14)
print(len(tweet_df_train), len(tweet_df_dev), len(tweet_df_test))
tweet_df_train = tweet_df_train.reset_index(drop=True)
tweet_df_dev = tweet_df_dev.reset_index(drop=True)
tweet_df_test = tweet_df_test.reset_index(drop=True)
tweet_df_train.to_pickle('data/tweet_train_orig.pkl')
tweet_df_dev.to_pickle('data/tweet_dev_orig.pkl')
tweet_df_test.to_pickle('data/tweet_test_orig.pkl')
```

```python
tweet_df_train['hashtags'] = tweet_df_train['Tweet'].apply(lambda x: re.findall(r'\B#\w*[a-zA-Z]+\w*', x)) 
tweet_df_dev['hashtags'] = tweet_df_dev['Tweet'].apply(lambda x: re.findall(r'\B#\w*[a-zA-Z]+\w*', x)) 
tweet_df_train['mentions'] = tweet_df_train['Tweet'].apply(lambda x: re.findall(r'\B@\w*[a-zA-Z]+\w*', x)) 
tweet_df_dev['mentions'] = tweet_df_dev['Tweet'].apply(lambda x: re.findall(r'\B@\w*[a-zA-Z]+\w*', x))
tweet_df_train.head()
```

```python
def preprocess_tweet(tweet, remove_emoji = True):
    tweet = tweet.replace('RT','')
    tweet = re.sub(r"http\S+", "", tweet) 
    tweet = re.sub(r"\s+"," ",tweet)
    tweet = re.sub(r"@\w+[:]?", "", tweet)

    clean = re.sub(r"""
               [,.;:@#?!&$]+  # Accept one or more copies of punctuation
               \ *           # plus zero or more copies of a space,
               """,
               " ",          # and replace it with a single space
               tweet, flags=re.VERBOSE)
    if remove_emoji:
        tweet = demoji.replace(tweet)
    else:
        tweet = demoji.replace_with_desc(tweet)
    tweet = tweet.strip()
    return tweet
```

```python
print('1)Orignal tweet: \n\n',tweet_df.iloc[0]['Tweet'])
print('\n2)With emoji descriptions: \n\n',preprocess_tweet(tweet_df.iloc[0]['Tweet'],remove_emoji=False))
print('\n3)Without emoji descriptions: \n\n',preprocess_tweet(tweet_df.iloc[0]['Tweet'],remove_emoji=True))
```

```python
tweet_df_train['Tweet_with_emoji_desc'] = tweet_df_train['Tweet'].apply(preprocess_tweet,remove_emoji = False)
tweet_df_train['Tweet'] = tweet_df_train['Tweet'].apply(preprocess_tweet)
tweet_df_dev['Tweet_with_emoji_desc'] = tweet_df_dev['Tweet'].apply(preprocess_tweet,remove_emoji = False)
tweet_df_dev['Tweet'] = tweet_df_dev['Tweet'].apply(preprocess_tweet)
tweet_df_train.head()
```

```python
tweet_df_train.to_pickle('data/tweet_train_cleaned.pkl')
tweet_df_dev.to_pickle('data/tweet_dev_cleaned.pkl')
```

```python
article_df = pd.read_excel('Development Data/dev_data_article.xlsx')

article_df_train_dev, article_df_test = train_test_split(article_df,stratify=article_df['Mobile_Tech_Flag'],random_state=21,test_size = 0.1)
article_df_train, article_df_dev  = train_test_split(article_df_train_dev,stratify=article_df_train_dev['Mobile_Tech_Flag'],random_state=21,test_size = 0.14)
print(len(article_df_train), len(article_df_dev), len(article_df_test))
article_df_train = article_df_train.reset_index(drop=True)
article_df_dev = article_df_dev.reset_index(drop=True)
article_df_test = article_df_test.reset_index(drop=True)
article_df_train.to_pickle('data/article_train_orig.pkl')
article_df_dev.to_pickle('data/article_dev_orig.pkl')
article_df_test.to_pickle('data/article_test_orig.pkl')
```

```python
# Storing all the urls in a different column
article_df_train['urls'] = article_df_train['Text'].apply(lambda x: re.findall(r"http\S+", x)) 
article_df_dev['urls'] = article_df_dev['Text'].apply(lambda x: re.findall(r"http\S+", x)) 
article_df_train.head()
```

```python
def preprocess_article(article):
    article = re.sub(r"http\S+", "", article)
    article = re.sub(r"www.\S+", "", article)
    article = re.sub('[.]\n?', ". ",article)
    article = re.sub('\n+', " ",article)
    article = article.strip()
    if len(article) > 60:
        article = article[:-60]
    return article
```

```python
article_df_train['Text'] = article_df_train['Text'].apply(preprocess_article)
article_df_dev['Text'] = article_df_dev['Text'].apply(preprocess_article)
article_df_train.head()
```

```python
article_df_train.to_pickle('data/article_train_cleaned.pkl')
article_df_dev.to_pickle('data/article_dev_cleaned.pkl')
```
